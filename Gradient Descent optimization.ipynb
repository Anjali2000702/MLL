```python
import numpy as np

# Sample dataset
X = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 4, 5])

# Initialize parameters
m = 0
c = 0

# Learning rate
alpha = 0.01

# Number of iterations
epochs = 1000

n = len(X)

for i in range(epochs):
    y_pred = m * X + c

    # Calculate gradients
    dm = (-2/n) * sum(X * (y - y_pred))
    dc = (-2/n) * sum(y - y_pred)

    # Update parameters
    m = m - alpha * dm
    c = c - alpha * dc

print("Final slope (m):", m)
print("Final intercept (c):", c)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.plot(X, m*X + c)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Gradient Descent Linear Regression")
plt.show()
```
